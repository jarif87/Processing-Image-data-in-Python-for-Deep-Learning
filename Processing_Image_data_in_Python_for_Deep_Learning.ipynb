# convert rgb to grayscale image

```python
import cv2
```

```python
imag2=cv2.imread("/content/cat.jpg")
```

```python
type(imag2)
```

```python
imag2.shape
```

```python
gray_image=cv2.cvtColor(imag2,cv2.COLOR_RGB2GRAY)
type(gray_image)
```

```python
gray_image.shape
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(gray_image)
plt.show()
```

```python
cv2.imwrite("gray_image_0.jpg",gray_image)
```
